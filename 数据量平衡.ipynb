### 随机删除图片，使得其他种类的样本量和样本量最少的种类相一致

```python
import os
import random

# 设置工作目录
root_dir = "E:/jupyter/VIT_example/dataset/hfdataset_Mini/train"

# 获取所有子文件夹
folders = [os.path.join(root_dir, folder) for folder in os.listdir(root_dir) if os.path.isdir(os.path.join(root_dir, folder))]

# 确定每个文件夹中的图片数量
folder_image_counts = {}
for folder in folders:
    images = [img for img in os.listdir(folder) if img.lower().endswith(('.png', '.jpg', '.jpeg'))]
    folder_image_counts[folder] = len(images)

# 找到最小图片数量
min_image_count = min(folder_image_counts.values())

# 删除多余的图片
for folder, count in folder_image_counts.items():
    if count > min_image_count:
        # 获取所有图片
        images = [img for img in os.listdir(folder) if img.lower().endswith(('.png', '.jpg', '.jpeg'))]
        # 随机选择要删除的图片
        images_to_delete = random.sample(images, count - min_image_count)
        # 删除图片
        for img in images_to_delete:
            os.remove(os.path.join(folder, img))
            print(f"Deleted {img} from {folder}")

print("complete")
```
