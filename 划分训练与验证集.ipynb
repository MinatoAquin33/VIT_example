```python
import os
import shutil
import random

def move_files_to_validation(root_dir, train_subdir, val_subdir, fraction=0.3):

    train_dir = os.path.join(root_dir, train_subdir)
    val_dir = os.path.join(root_dir, val_subdir)
    
    os.makedirs(val_dir, exist_ok=True)
    
    # 
    files = [f for f in os.listdir(train_dir) if os.path.isfile(os.path.join(train_dir, f))]
    
    # 
    num_files_to_move = int(len(files) * fraction)
    
    # 
    files_to_move = random.sample(files, num_files_to_move)
    
    # 
    for file in files_to_move:
        src_path = os.path.join(train_dir, file)
        dst_path = os.path.join(val_dir, file)
        shutil.move(src_path, dst_path)
        print(f"Moved {file} from {train_dir} to {val_dir}")

# 
root_directory = "E:/jupyter/VIT_example/dataset/hfdataset_Mini"
train_subdirs = ["train/Covid", "train/Lung Opacity", "train/Normal", "train/Viral Pneumonia"]
val_subdirs = ["test/Covid", "test/Lung Opacity", "test/Normal", "test/Viral Pneumonia"]

# 
for train_subdir, val_subdir in zip(train_subdirs, val_subdirs):
    move_files_to_validation(root_directory, train_subdir, val_subdir)
```
