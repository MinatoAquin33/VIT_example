# 扫描文件并创建列表

```python
import os
# 初始化文件列表
file_list = []

# 指定根目录
root_dir = 'E:/jupyter/VIT_example/dataset/hfdataset_Full/train'

# 遍历目录
for dirpath, dirnames, filenames in os.walk(root_dir):
    for filename in filenames:
        # 将文件的完整路径添加到列表中
        file_path = os.path.join(dirpath, filename)
        file_list.append(file_path)
```

```python
print(file_list[0])
```

# 随机旋转图像

```python
from PIL import Image
import random

# 读取file_list中的文件，将这些文件随机旋转0-360度
for file_path in file_list:
    # 打开图像
    with Image.open(file_path) as img:
        # 生成随机旋转角度
        angle = random.randint(0, 360)
        # 旋转图像
        rotated_img = img.rotate(angle, expand=True)  # expand=True 使图像扩展到足以容纳整个旋转图像

        # 构建新的文件名
        base, ext = os.path.splitext(file_path)
        new_filename = f"{base}_RR{ext}"

        # 保存旋转后的图像
        rotated_img.save(new_filename)
        print(f"Saved rotated image {new_filename}")

```

# 水平翻转图像

```python
from PIL import Image
import random

# 读取file_list中的文件，将这些文件随机旋转0-360度
for file_path in file_list:
    # 打开图像
    with Image.open(file_path) as img:
        # 水平翻转图像
        flipped_img = img.transpose(Image.FLIP_LEFT_RIGHT)

        # 构建新的文件名
        base, ext = os.path.splitext(file_path)
        new_filename = f"{base}_HF{ext}"

        # 保存旋转后的图像
        flipped_img.save(new_filename)
        print(f"horizontally flipped image {new_filename}")
```

# 亮度增强

```python
from PIL import Image, ImageEnhance
import random

# 读取file_list中的文件，将这些文件随机旋转0-360度
for file_path in file_list:
    # 打开图像
    with Image.open(file_path) as img:
        
        # 调整亮度
        enhancer = ImageEnhance.Brightness(img)
        min_factor = 0.5
        max_factor = 1.5
        factor = random.uniform(min_factor, max_factor)
        img_enhanced = enhancer.enhance(factor)
        

        # 构建新的文件名
        base, ext = os.path.splitext(file_path)
        new_filename = f"{base}_BR{ext}"

        # 保存调整亮度后的图像
        img_enhanced.save(new_filename)
        print(f"brightness adjusted image {new_filename}")
```

# 对比度增强

```python
from PIL import Image, ImageEnhance
import random

# 读取file_list中的文件，将这些文件随机旋转0-360度
for file_path in file_list:
    # 打开图像
    with Image.open(file_path) as img:
        
        # 调整亮度
        enhancer = ImageEnhance.Contrast(img)
        min_factor = 0.5
        max_factor = 1.5
        factor = random.uniform(min_factor, max_factor)
        img_enhanced = enhancer.enhance(factor)
        

        # 构建新的文件名
        base, ext = os.path.splitext(file_path)
        new_filename = f"{base}_CT{ext}"

        # 保存调整亮度后的图像
        img_enhanced.save(new_filename)
        print(f"contrast adjusted image {new_filename}")
```
